```python
import cv2

IDX = 12  # <-- поменяй на свой
cap = cv2.VideoCapture(IDX)

while True:
    ret, frame = cap.read()
    if not ret:
        print("No frame")
        break

    cv2.imshow("Camera", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

working = []
for i in range(0, 20):
    cap = cv2.VideoCapture(i)
    ok = cap.isOpened()
    ret, frame = (False, None)
    if ok:
        ret, frame = cap.read()
    cap.release()

    if ok and ret and frame is not None:
        print(f"✅ index {i} works, shape={frame.shape}")
        working.append(i)
    else:
        print(f"❌ index {i} no")

print("WORKING:", working)
```
