```python
import cv2

working = []

for i in range(0, 20):
    cap = cv2.VideoCapture(i)
    opened = cap.isOpened()

    ret, frame = (False, None)
    if opened:
        ret, frame = cap.read()

    cap.release()

    if opened and ret and frame is not None:
        print(f"✅ INDEX {i} WORKS, frame shape = {frame.shape}")
        working.append(i)
    else:
        print(f"❌ index {i} does NOT work")

print("\nWORKING INDEXES:", working)
```

```python
import cv2
from matplotlib import pyplot as plt

IDX = 0  # <-- ваш индекс
cap = cv2.VideoCapture(IDX)
ret, frame = cap.read()
cap.release()

print("ret:", ret, "shape:", None if frame is None else frame.shape)

# OpenCV даёт BGR, переводим в RGB для matplotlib
rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
plt.axis("off")
plt.show()
```

```python
import cv2

IDX = 0  # <-- поменяй на свой
cap = cv2.VideoCapture(IDX)

while True:
    ret, frame = cap.read()
    if not ret:
        print("No frame")
        break

    cv2.imshow("Camera", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
