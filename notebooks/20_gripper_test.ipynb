```python
from control.gripper import GripperController, GripperConfig

# ВАЖНО: сначала тестируем БЕЗ движения
config = GripperConfig(
    motor_name="m6",   # если не уверены — проверим ниже
    open_angle=30,
    close_angle=90,
    speed=40
)

gripper = GripperController(config=config, dry_run=True)

gripper.connect()

gripper.poppy.rest_posture.start()

gripper.poppy.m1.compliant = False
gripper.poppy.m2.compliant = False
gripper.poppy.m3.compliant = False
gripper.poppy.m4.compliant = False
gripper.poppy.m5.compliant = False
gripper.poppy.m6.compliant = False

print("Connected. Available motors:", gripper.list_motor_names())
```

```python
gripper.open()
gripper.close()
```

```python
# ВНИМАНИЕ: теперь мотор БУДЕТ двигаться
gripper.dry_run = False

print("Opening gripper...")
gripper.open()

import time
time.sleep(2)

print("Closing gripper...")
gripper.close()
```

```python
# Подбираем углы вручную
for angle in [20, 30, 40, 50]:
    print("Trying angle:", angle)
    gripper._move_to(angle, label="TEST")
    time.sleep(1)
```
