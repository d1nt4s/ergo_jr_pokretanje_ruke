```python
while True:
    ret, frame = cap.read()
    if not ret:
        continue

    raw = detector.infer(frame)
    confirmed = deb.update(raw)

    if confirmed == "OPEN":
        print("OPEN detected")
        gripper.open()

    elif confirmed == "CLOSE":
        print("CLOSE detected")
        gripper.close()

    time.sleep(0.02)
```
